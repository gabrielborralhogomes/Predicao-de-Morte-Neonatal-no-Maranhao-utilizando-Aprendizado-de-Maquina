```python
# Definindo as colunas para comparação
colunas_para_comparar = [
    'DTNASC',
    'SEXO',
    'CODMUNNASC',
    'IDADEMAE',
    'RACACOR',
    'ESCMAE'
]

# Fazendo uma cópia do df_sinasc e adicionando a coluna "VIVO" preenchida com 1
df_neom = df_sinasc.copy()  # Usando df_neom agora
df_neom['VIVO'] = 1

# Inicializando os contadores
cont_acho = 0
cont_dup = 0

# Iterando sobre cada linha do df_sim
for index, row_sim in df_sim.iterrows():
    # Verificando se a linha do df_sim está presente no df_neom
    match = df_neom[
        (df_neom['DTNASC'] == row_sim['DTNASC']) &
        (df_neom['SEXO'] == row_sim['SEXO']) &
        (df_neom['CODMUNNASC'] == row_sim['CODMUNNASC']) &
        (df_neom['IDADEMAE'] == row_sim['IDADEMAE']) &
        (df_neom['RACACOR'] == row_sim['RACACOR']) &
        (df_neom['ESCMAE'] == row_sim['ESCMAE'])
    ]

    # Se houver correspondência
    if not match.empty:
        cont_acho += 1  # Incrementa o contador de ocorrências
        
        # Verificando se há duplicatas (mais de uma ocorrência)
        if match.shape[0] > 1:
            # Excluir todas as duplicatas do df_neom
            df_neom.drop(match.index, inplace=True)
            cont_dup += 1  # Incrementa o contador de duplicatas
        else:
            # Alterar o valor da coluna "VIVO" para 0 na ocorrência encontrada
            df_neom.loc[match.index, 'VIVO'] = 0

# Exibindo os resultados
print(f"Número de ocorrências encontradas: {cont_acho}")
print(f"Número de duplicatas encontradas: {cont_dup}")
```
